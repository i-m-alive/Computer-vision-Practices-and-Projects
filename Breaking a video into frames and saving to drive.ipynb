```python
import cv2
```

```python
path = input("Enter the path of the video from the disk: ")
vidcap = cv2.VideoCapture(path)
#capture the video in the *vidcap* object

ret, img = vidcap.read()  # Read the first frame of the video
count = 0

while ret:
    cv2.imwrite("D:\\frames\\img%d.jpg" % count, img)
    #saving the image using imwrite
    #why we use %d and %count here?
    # as we wanted to save each frame diffrent name but easy to see the order of the frames
    milliseconds_per_frame = 1000 / vidcap.get(cv2.CAP_PROP_FPS)
    # Calculate the duration of each frame in milliseconds
    vidcap.set(cv2.CAP_PROP_POS_MSEC, count * milliseconds_per_frame)
    #set(): This method is used to set properties of the video capture object.
    #cv2.CAP_PROP_POS_MSEC: This constant represents the current position of the video file in milliseconds.

    ret, img = vidcap.read()# Read the next frame
    cv2.imshow('live',img)
    count +=1
    print(count)
    if cv2.waitKey(1)==ord('q'):
        break
          
vidcap.release()
cv2.destroyAllWindows()
        
```
